## Guessing Game Challenge :

Let's use while loops to create a guessing game
e:

Write a program that picks a random integer from 1 to 100, and has players guess the number. The rulesare:

If a player's guess is less than 1 or greater than 100, say "OUT OF BOUNDS"
On a player's first turn, if their guess is
within 10 of the number, return "WARM!"
further than 10 away from the number, return "COLD!"
On all subsequent turns, if a guess is
closer to the number than the previous guess return "WARMER!"
farther from the number than the previous guess, return "COLDER!"
When the player's guess equals the number, tell them they've guessed correctly and how many guesses it took!
You can try this from scratch, or follow the steps outlined below. A separate Solution notebook has been provided. Good luck!

```python
import random
num = random.randint(1,100)
guesses = [0]


print("WELCOME TO GUESS ME!")
print("I'm thinking of a number between 1 and 100")
print("If your guess is more than 10 away from my number, I'll tell you you're COLD")
print("If your guess is within 10 of my number, I'll tell you you're WARM")
print("If your guess is farther than your most recent guess, I'll say you're getting COLDER")
print("If your guess is closer than your most recent guess, I'll say you're getting WARMER")
print("LET'S PLAY!")


import random  # Needed if you want to generate a random number instead of a fixed one

num = 42  # The number the user has to guess (can use random.randint(1,100) for randomness)
guesses = [0]  # List to store all guesses; start with 0 as a placeholder

while True:
    # Ask the user for their guess
    guess = int(input("I'm thinking of a number between 1 and 100.\n  What is your guess? "))
    
    # Check if the guess is outside the valid range
    if guess < 1 or guess > 100:
        print('OUT OF BOUNDS! Please try again: ')
        continue  # Skip the rest of the loop and ask again

    # Check if the guess is correct
    if guess == num:
        print(f'CONGRATULATIONS, YOU GUESSED IT IN ONLY {len(guesses)} GUESSES!!')
        break  # Exit the loop since the user has guessed correctly

    # Store the guess for future comparisons
    guesses.append(guess)

    # Compare the current guess with the previous guess to give feedback
    if guesses[-2]:  # If there is a previous guess (not the first one)
        if abs(num - guess) < abs(num - guesses[-2]):
            print('WARMER!')  # Current guess is closer than previous
        else:
            print('COLDER!')  # Current guess is farther than previous
    else:
        # For the first real guess, give WARM or COLD feedback based on proximity
        if abs(num - guess) <= 10:
            print('WARM!')
        else:
            print('COLD!')
```
